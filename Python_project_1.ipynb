```python
#Merging all the sheets in excel into one dataframe
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df=pd.concat(pd.read_excel("C:\\Users\\judaybhx\\OneDrive - Intel Corporation\Desktop\\Data Science\\imus+(2).xlsx",sheet_name=None),ignore_index=True)
df
```

```python
#separating only stock products
df_stock=df[df['status']=='In Stock']
df_stock
```

```python
#separating only No stock products
df_Nostock=df[df['status']=='No Stock']
df_Nostock
```

```python
#1. Which are the top Manufacturers with price, quantity?
m=df_stock.groupby(['price','qty'])['manufacturer']
m=pd.DataFrame(m)
m.tail()
#it is not a exact value i think 
```

```python
#2. Which are the most profitable products?
df_stock.sort_values('price',ascending=False)
```

```python
#3. Which is the product where you have seen a decrease in the price?
d=df_stock.groupby(['part_number','sku_create_date'])['price'].max()
d=pd.DataFrame(d)
d.sort_values(by='price',ascending=False)
```

```python
#4. Which is the product where you have seen an increase in the price?
d=df_stock.groupby(['part_number','sku_create_date'])['price'].min()
d=pd.DataFrame(d)
d.sort_values(by='price')
```

```python
#5. Which is the top category by price and quantity?
d=df_stock.groupby(['price','qty'])['category'].max()
d=pd.DataFrame(d)
print(d)
```
